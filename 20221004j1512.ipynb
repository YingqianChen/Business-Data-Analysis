```python
import requests
from bs4 import BeautifulSoup



glxynewsurl="https://sm.xmu.edu.cn/info/1049/32764.htm"
glxyres=requests.get(glxynewsurl)

glxyres.encoding="UTF8"
soup11=BeautifulSoup(glxyres.text,'html.parser')

datainfolist=soup11.select('p')
for datainfo in datainfolist:
  print(datainfo.text)
```

```python
spaninfolist=soup11.select('span')
for spandata in spaninfolist:
  print(spandata.text)
```

```python
spaninfolist1=soup11.select('div span')
for spandata in spaninfolist1:
  print(spandata.text)
```
