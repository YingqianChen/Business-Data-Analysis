```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams['font.sans-serif'] = 'SimHei'
```

```python
population = np.load('populations.npz', allow_pickle=True)
data = population["data"]
feature = population["feature_names"]
data = pd.DataFrame(data)
data = data.drop(data.tail(2).index)
data = data.sort_values(by = 0)
```

```python
p = plt.figure(figsize = (12, 12))
ax1 = p.add_subplot(2,1,1)
plt.scatter(data[0], data[1])
plt.scatter(data[0], data[2], marker = '8')
plt.scatter(data[0], data[3], marker = 'D')
plt.scatter(data[0], data[4], marker = 'v')
plt.scatter(data[0], data[5], marker = 'p')
plt.xlabel("年份")
plt.ylabel("人数（万）")
plt.title('1996-2015年人口数据特征散点图')
plt.legend(feature[1:])
?
ax2 = p.add_subplot(2,1,2)
plt.plot(data[0], data[1], 'r-',
        data[0], data[2], 'b-',
        data[0], data[3], 'y--',
        data[0], data[4], 'c-',
        data[0], data[5], 'k--')
plt.xlabel("年份")
plt.ylabel("人数（万）")
plt.title('1996-2015年人口数据特征折线图')
plt.savefig('1996-2015年人口数据特征.png')
plt.legend(feature[1:])
plt.show()
```

## 分析未来人口趋势
- 年末总人口、男性人口、女性人口、城镇人口均呈现上升趋势
- 乡村人口呈现下降趋势
- 增长最快的是城镇人口

```python
p = plt.figure(figsize = (12,12))
ax1 = p.add_subplot(2,2,1)
plt.bar(data[0], data[2])
plt.xticks(rotation=90, fontsize=10)
plt.xlabel("年份")
plt.ylabel("人数（万）")

ax2 = p.add_subplot(2,2,2)
plt.bar(data[0], data[3])
plt.xticks(rotation=90, fontsize=10)
plt.xlabel("年份")
plt.ylabel("人数（万）")

ax3 = p.add_subplot(2,2,3)
plt.bar(data[0], data[4])
plt.xticks(rotation=90, fontsize=10)
plt.xlabel("年份")
plt.ylabel("人数（万）")

ax4 = p.add_subplot(2,2,4)
plt.bar(data[0], data[5])
plt.xticks(rotation=90, fontsize=10)
plt.xlabel("年份")
plt.ylabel("人数（万）")

plt.savefig('1996-2015年人口数据特征分布情况1.png')
plt.show()
```

```python
label1 = ['男性人口', '女性人口']
label2 = ['城镇人口', '乡村人口']
explode = [0.02, 0.02]

p = plt.figure(figsize = (8, 8))
ax1 = p.add_subplot(2,2,1)
plt.pie((data[2][19], data[3][19]), autopct = '%.2f%%', labels = label1, explode = explode)
plt.title('1996年男女人口比例')

ax2 = p.add_subplot(2,2,2)
plt.pie((data[2][0], data[3][0]), autopct = '%.2f%%', labels = label1, explode = explode)
plt.title('2015年男女人口比例')

ax3 = p.add_subplot(2,2,3)
plt.pie((data[4][19], data[5][19]), autopct = '%.2f%%', labels = label2, explode = explode)
plt.title('1996年城乡人口比例')

ax4 = p.add_subplot(2,2,4)
plt.pie((data[4][0], data[5][0]), autopct = '%.2f%%', labels = label2, explode = explode)
plt.title('2015年城乡人口比例')

plt.savefig('1996-2015年人口数据特征分布情况2.png')
plt.show()
```

```python
p = plt.figure(figsize = (10, 10))
data1 = ([list(data[i]) for i in range(1, 6)])
plt.boxplot(data1, notch = True, labels = feature[1:], meanline = True)
plt.xlabel('特征')
plt.ylabel('人数（万）')
plt.title('1996-2015年人口特征分布情况箱线图')

plt.savefig('1996-2015年人口数据特征分布情况3.png')
plt.show()
```

## 分析人口结构变化情况以及变化速率的增减状况
- 条形图及饼图
    - 1996年-2015年，除了乡村人口，其他人口特征均呈现增长趋势
    - 男性人口比例增长了0.6%，女性人口比例对应减少了0.6%
    - 城镇人口比例增长迅速，增长比例约为26%
- 箱线图
    - 整体人口数量是上升的
    - 各人口特征的增长都比较平缓
