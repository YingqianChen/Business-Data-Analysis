```python
import requests
from lxml import html
url_dbmusic='https://music.douban.com/' #需要爬的网址
header_db={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:54.0) Gecko/20100101 Firefox/54.0','Connection':'keep-alive'}
page_dbmusic=requests.get(url_dbmusic,headers=header_db) 
tree_dbmusic=html.fromstring(page_dbmusic.text) 
result_dbmusic=tree_dbmusic.xpath('//tr//a/text()') #需要获取的数据
result1_dbmusic=tree_dbmusic.xpath('//tr//a/@href')
result2_dbmusic=tree_dbmusic.xpath('//tr[last()]//a/@href')
print(result_dbmusic)
print(result1_dbmusic)
print(result2_dbmusic)
```
